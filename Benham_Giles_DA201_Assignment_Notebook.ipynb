Determine the number of new Covid cases in the UK over time

```python
# Access NHS API - Code copied from NHS developers guide
# https://coronavirus.data.gov.uk/details/developers-guide/main-api
from requests import get


def get_data(url):
    response = get(url, timeout=10)
    
    if response.status_code >= 400:
        raise RuntimeError(f'Request failed: { response.text }')
        
    return response.json()
    

if __name__ == '__main__':
    endpoint = (
        'https://api.coronavirus.data.gov.uk/v1/data?'
        'filters=areaType=nation;areaName=england&'
        'structure={"date":"date","newCasesBySpecimenDate":"newCasesBySpecimenDate"}'
    )
    
    data = get_data(endpoint)
    
    # Retrieve the data
    cases_by_date = data.get('data', [])

    # Create a pandas DataFrame using the Covid data
    cases = pd.DataFrame(cases_by_date)

    print(cases)
```

```python
# Convert the 'date' column data type to datetime
cases['date'] = pd.to_datetime(cases['date'])

# Extract the month and year from the 'date' column
cases['month'] = cases['date'].dt.to_period('M')

# Group the DataFrame by 'month' and calculate the sum of new cases per month
sum_cases = cases.groupby('month')['newCasesBySpecimenDate'].sum().reset_index()

# Rename the DataFrames columns
sum_cases.columns = ['month', 'sum_new_cases']

# Filter months column to line up with the other graphs
sum_cases = sum_cases[sum_cases['month'] <= '2022-06']

# Convert month column back to datetime
sum_cases['month'] = sum_cases['month'].dt.to_timestamp()

# Reset the index
sum_cases.reset_index(drop=True, inplace=True)

# View the DataFrame
sum_cases.head()
```

```python
# Create a line plot to show Covid cases over time
ax = sns.lineplot(data=sum_cases, x='month', y='sum_new_cases')

# Specify the labels
ax.set_xlabel('Month')
ax.set_ylabel('Sum of New Cases (Millions)')
ax.set_title('Montlhy New COVID Cases')
```
